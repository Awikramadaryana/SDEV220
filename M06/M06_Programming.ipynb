```python
import multiprocessing
import random
import time
from datetime import datetime
```

```python
def worker_process(process_id):
    """
    Worker function that each process will execute.
    
    Args:
        process_id: Unique identifier for the process
    """
    # Generate random wait time between 0 and 1 second
    wait_time = random.uniform(0, 1)
    
    # Wait for the random duration
    time.sleep(wait_time)
    
    # Get current time
    current_time = datetime.now().strftime("%H:%M:%S.%f")[:-3]  # Format with milliseconds
    
    # Print process information
    print(f"Process {process_id} waited {wait_time:.3f} seconds and exited at {current_time}")

def main():
    # Create a list to hold the process objects
    processes = []
    
    # Create and start 3 processes
    for i in range(3):
        process = multiprocessing.Process(target=worker_process, args=(i+1,))
        processes.append(process)
        process.start()
        print(f"Started process {i+1}")
    
    # Wait for all processes to complete
    for process in processes:
        process.join()
    
    print("\nAll processes have completed.")

if __name__ == "__main__":
    # This is important for Windows compatibility and proper multiprocessing behavior
    main()
```
